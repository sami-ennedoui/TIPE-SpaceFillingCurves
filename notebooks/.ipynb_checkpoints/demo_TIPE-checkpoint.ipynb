#  Space-Filling Curves & Image Processing

This notebook demonstrates the core concept of my project: using fractal space-filling curves (like the **Hilbert Curve**) to optimize image dithering and lossless compression.

### The Mathematical Concept
Standard image processing algorithms traverse pixels linearly (left-to-right, top-to-bottom). However, a Hilbert curve provides a continuous fractal bijection $ f: [0,1] \to [0,1]^2 $ that preserves 2D spatial locality. 
By forcing an error-diffusion algorithm (Atkinson) to follow this fractal path, we group similar pixels together, drastically improving the efficiency of run-length encoding compressors like GZIP.

```python
import sys
import os
import matplotlib.pyplot as plt

# Add the parent directory to the path so we can import our main pipeline
sys.path.append(os.path.abspath('..'))

# Import functions from our clean software architecture
from main_pipeline import (
    load_grayscale_image,
    hilbert_path,
    raster_path,
    atkinson_dither_grayscale,
    gzip_size_bytes_of_raw_u8
)

# Matplotlib configuration for nice HD plots
plt.rcParams['figure.dpi'] = 150
plt.rcParams['image.cmap'] = 'gray'
```

## 1. Loading the Source Image
We load a standard test image (8-bit grayscale) to evaluate our pipeline.

```python
# Load the image (ensure 'hopper.png' is in 'data/input/')
image_path = "../data/input/hopper.png"
original_img = load_grayscale_image(image_path, size=(256, 256))

h, w = original_img.shape
print(f"Original Image Shape: {w}x{h} pixels")

plt.figure(figsize=(4, 4))
plt.title("Original 8-bit Grayscale Image")
plt.axis("off")
plt.imshow(original_img)
plt.show()
```

```python
## 2. Generating Paths & Applying Atkinson Dithering
We will now convert this 8-bit image (256 shades of gray) into a 1-bit image (black and white only).
To do this, we use the **Atkinson dithering algorithm**, but we evaluate two different spatial traversal methods:
1. **Raster Path:** The standard linear scan.
2. **Hilbert Path:** Our optimized fractal scan.
```

```python
# 1. Generate coordinates sequences
path_hilbert = hilbert_path(w, h)
path_raster = raster_path(w, h)

# 2. Apply dithering along the specific paths
dithered_hilbert = atkinson_dither_grayscale(original_img, path_hilbert)
dithered_raster = atkinson_dither_grayscale(original_img, path_raster)

print(" Dithering completed successfully.")
```

## 3. Visual Comparison
Let's look at the results. At a macro level, both images simulate grayscale well. However, the spatial structure of the pixels is entirely different. The Hilbert curve creates continuous "clusters" of black and white pixels, reducing high-frequency noise.


```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(dithered_raster)
axes[0].set_title("Standard Raster Dithering (1-bit)")
axes[0].axis("off")

axes[1].imshow(dithered_hilbert)
axes[1].set_title("Hilbert Curve Dithering (1-bit)")
axes[1].axis("off")

plt.tight_layout()
plt.show()
```

## 4. Quantitative Results: Lossless Compression
Because the Hilbert curve preserves spatial locality, standard compression algorithms (like GZIP, which relies on the Lempel-Ziv algorithm) can compress the binary stream much more effectively. Let's measure the exact file sizes.


```python
# Calculate compressed sizes
size_raster = gzip_size_bytes_of_raw_u8(dithered_raster)
size_hilbert = gzip_size_bytes_of_raw_u8(dithered_hilbert)

# Calculate global reduction compared to the raw 8-bit image
raw_size = w * h  # 1 byte per pixel originally
reduction_hilbert = 100.0 * (raw_size - size_hilbert) / raw_size

print(" Compression Benchmark (GZIP):")
print("-" * 40)
print(f"Raw 8-bit original size : {raw_size} bytes")
print(f"Raster dithered size    : {size_raster} bytes")
print(f"Hilbert dithered size   : {size_hilbert} bytes")
print("-" * 40)

gain_vs_raster = 100.0 * (size_raster - size_hilbert) / size_raster
print(f" Algorithmic gain (Hilbert vs Raster) : -{gain_vs_raster:.2f}%")
print(f" Global Pipeline Size Reduction       : -{reduction_hilbert:.2f}%")
```
